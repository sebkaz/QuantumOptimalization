```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rna = pd.read_csv('genus0prote.csv', sep=';')
```

```python
wszystkie_rna = len(rna)
mean_rna = rna['sequence length'].mean()
median_rna = rna['sequence length'].median()
```

```python
print(f"Struktury białek o genusie 0: {wszystkie_rna} ")
print(f"średnia długość łańcuchów: {mean_rna}")
print(f"mediana długości łańcuchów: {median_rna}")
```
